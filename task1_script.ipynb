```python
#Download dataset

import requests  

print('Download Starting...')
url = "https://data.gov.sg/dataset/21d477f2-6e1b-4232-82b3-59e804dc2f6a/download"
r = requests.get(url)
#filename = url.split('/')[-1]
with open("download.zip",'wb') as output_file:
    output_file.write(r.content)
print('Download Completed!!!')
```

```python
#Extract zipfile

from zipfile import ZipFile
file = "download.zip" 
with ZipFile(file, 'r') as zip:
    zip.printdir()
    print('Extracting all the files now...')
    zip.extractall(path=None, members=None, pwd=None)
    print('Done!')
```

```python
#read multiple csv - collect file names

import pandas as pd
import glob

path = ""
all_files = glob.glob(path + "*.csv")
all_files
```

```python
#read multiple csv - for loop

li = []
for filename in all_files:
    df = pd.read_csv(filename, delimiter=',')
    li.append(df)

li[1]
df = pd.concat(li)
df
```

```python
#Rename entity_name to Company Name
df.rename(columns={'entity_name':'Company Name'}, inplace = True)
df
```

```python
#Create finaladdress column by combining block,	street_name, building_name,	level_no,unit_no, postal_code

#remove na
df['building_name1'] = df['building_name'].replace('na','')
df['level_no1'] = '#' + df['level_no'].replace('na','')
df['unit_no1'] = df['unit_no'].replace('na','')

#concatenate address
df['address1'] = df['block'] + ' ' + df['street_name'] + ' ' + df['building_name1']
df['address2'] = df['level_no1'].str.cat(df['unit_no1'],sep = "-", na_rep="")
df['finaladdress'] = df['address1'] + ' ' + df['address2'].replace('#-','') + ' Singapore ' + df['postal_code'].astype(str) + ' '
df['finaladdress']
df
```

```python
#Clean date uen_issue_date	registration_incorporation_date

import datetime

df['uen_issue_date']=pd.to_datetime(df['uen_issue_date'])
df['registration_incorporation_date']=pd.to_datetime(df['registration_incorporation_date']).dt.floor('d')
df['registration_incorporation_date']
```

```python
#Add new column ACRA Update Date
from datetime import date
df['ACRA Update Date'] = date.today()
```

```python
#Create a new df using selected columns in csv

df_new = df[["uen","Company Name","primary_ssic_code","secondary_ssic_code","primary_ssic_description","secondary_ssic_description","finaladdress","entity_status_description","uen_issue_date","registration_incorporation_date","entity_type_description","ACRA Update Date"]]

df_new
```

```python
#Read UEN list

df_uen = pd.read_csv('task1_uen_list.csv')
df_uen.head(5)
```

```python
#merge file

df3 = pd.merge(df_new,df_uen,on = 'uen', how = 'inner')
df3
```
