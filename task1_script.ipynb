```python
#Download dataset

import requests  

print('Download Starting...')
url = "https://data.gov.sg/dataset/21d477f2-6e1b-4232-82b3-59e804dc2f6a/download"
r = requests.get(url)
#filename = url.split('/')[-1]
with open("download.zip",'wb') as output_file:
    output_file.write(r.content)
print('Download Completed!!!')
```

```python
#Extract zipfile

from zipfile import ZipFile
file = "download.zip" 
with ZipFile(file, 'r') as zip:
    zip.printdir()
    print('Extracting all the files now...')
    zip.extractall(path=None, members=None, pwd=None)
    print('Done!')
```

```python
#read multiple csv - collect file names

import pandas as pd
import glob

#path = ""
path = "data/"
all_files = glob.glob(path + "*.csv")
all_files
```

```python
#read multiple csv - for loop

li = []
for filename in all_files:
    df = pd.read_csv(filename, delimiter=',')
    li.append(df)

li[1]
df = pd.concat(li)
df.head(5)
```

```python

df.shape
```

```python
#Create a new df using selected columns in csv

df_new = df[["uen","entity_name","primary_ssic_code","secondary_ssic_code","primary_ssic_description","secondary_ssic_description","block","street_name","building_name","level_no","unit_no","postal_code","entity_status_description","uen_issue_date","registration_incorporation_date","entity_type_description"]]

df_new.shape
```

```python
#Read UEN list

df_uen = pd.read_csv('task1_uen_list.csv')
#df_uen.shape
df_uen.head(5)
```

```python
#merge file
#https://statisticsglobe.com/types-joins-for-pandas-dataframes-python

df3 = pd.merge(df_new,df_uen,on = 'uen', how = 'inner')
#df3.shape
df3.head(5)
```

```python
#export to csv
df3.to_csv('task1_output.csv', sep=',',index=False)
```
